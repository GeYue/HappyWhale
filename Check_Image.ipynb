```python
import albumentations as A
from albumentations.pytorch import ToTensorV2

scale_size = 768
transform = A.Compose([
    A.LongestMaxSize(max_size=scale_size, interpolation=1),
    A.PadIfNeeded(min_height=scale_size, min_width=scale_size, border_mode=0, value=(0,0,0)),
    #A.Resize(height=scale_size, width=scale_size, interpolation=1),
    A.HueSaturationValue(hue_shift_limit=2, sat_shift_limit=3, val_shift_limit=2, p=1),
    A.RandomBrightnessContrast(p=1),    
    #A.RandomGamma(p=1),
    
    #A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0),
    #A.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5], max_pixel_value=255.0, p=1.0),    

    #A.CoarseDropout(max_holes=20, max_height=20, max_width=20, fill_value=0, p=1),
    #A.ToGray(p=1),    
])

img = np.array(img)
transformed = transform(image=img)

plt.figure(figsize=(20,20))
plt.imshow(transformed['image'])
plt.show()
```
